```python
import numpy as np
from sklearn.linear_model import LinearRegression # importa a regressão linear
import matplotlib.pyplot as plt # importa a biblioteca matplotlib
# Dados de treinamento
vazao = np.array([10,20,30,5,35,40]).reshape(-1,1) # vazao de agua em m3/min
nivel = np.array([50,60,70,45,80,85]) # nivel do tanque em m

# Criar e treinar o modelo de regressão linear
modelo = LinearRegression()
modelo.fit(vazao,nivel)  # treina o modelo
# Dados de teste
vazao_teste=np.array([15,25,33]).reshape(-1,1) # vazao de agua em m3/min
```

```python
# Dados de teste
#vazao_teste=np.array([15,25,33]).reshape(-1,1) # vazao de agua em m3/min
vazao_teste = np.array([5,2,7,30,50]).reshape(-1,1)

# Fazer previsoes com base na vazao de agua
nivel_previsto = modelo.predict(vazao_teste)
nivel_previsto
```

```python
# imprimir as previsões
for i in range(len(vazao_teste)):
    print(f"Vazao: {vazao_teste[i][0]} m3/min - Nível previsto: {nivel_previsto[i]} metros")

plt.scatter(vazao,nivel,color='blue',label='Dados de treinamento')
plt.plot(vazao_teste,nivel_previsto,color='red',label='Dados de teste')
plt.xlabel('Vazão de água (m3/min)')
plt.ylabel('Nivel do tanque (metros)')
plt.title('Relação entre Vazão de Água e Nível do Tanque')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Exercício 2 - Regressão polimonial
import numpy as np # importa a bibliote numpy
import matplotlib.pyplot as plt # importa a biblioteca matplotlib
from sklearn.linear_model import LinearRegression # importa a regressão linear
from sklearn.preprocessing import PolynomialFeatures # importa a regressão polinomial
```

```python
# Dados fornecidos
Q = np.array([100,200,300,40,500,600,700]).reshape(-1,1) # quantidade de calor
T = np.array([30,35,40,45,60,100,150]) # temperatura
```

```python
print(Q)
print(T)
```

```python
# Modelo 1 Regressão Linear simples
linear_model = LinearRegression() # cria o modelo de regressão linear
linear_model.fit(Q,T) # treina o modelo
T_linear_pred = linear_model.predict(Q) # realiza predições de temparatura com regressão linear
print(T_linear_pred)
```

```python
# Modelo 2 Regressão polinomial grau 2
poly = PolynomialFeatures(degree=2) # cria o modelo de regressão polinomial grau 2
Q_poly = poly.fit_transform(Q) # treina o modelo polinomial
poly_model = LinearRegression() # armazena modelo
poly_model.fit(Q_poly,T) # armazena modelo
```

```python
# Geração de valores para curvas
Q_range = np.linspace(100,700,500).reshape(-1,1)
T_linear_curve = linear_model.predict(Q_range) # curva linear
t_poly_curve = poly_model.predict(poly.fit_transform(Q_range)) # curva polinomial
# plotagem do gráfico com os modelos de machine learning
plt.figure(figsize=(10,6))
plt.scatter(Q,T,color='black',label='Dados reais')
plt.plot(Q_range,T_linear_curve,color='blue',label='Regressão Linear Simples')
plt.plot(Q_range,t_poly_curve,color='blue',linestyle='--', label='Regressão Polinomial Grau 2')
plt.title('Previsão de temperatura do Forno com regressão')
plt.xlabel('Quantidade de calor (Q)')
plt.ylabel('Temperatura (T)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
